```python
from serial import Serial
import time
import random
import json
import numpy as np
```

```python
letters = ['A', 'B', 'C', 'D', 'E']
```

```python
ser = Serial('/dev/ttyACM2',9600)

Gyrx = [[]]
Gyry = [[]]
Gyrz = [[]]
Accx = [[]]
Accy = [[]]
Accz = [[]]
labels = []

sample_size = 200
min_length = 50
steady_hand_threshold = 500
steady_hand_window = 20

count = 0

# Just a sleeping time to pick up the pen and hold it in the right position.
time.sleep(5)

print("GO")

letra = random.choice(letters)
print("0 - " + letra)
letter_counter = 0

while count < sample_size:
    try:
        ser.write('.'.encode())
        arduinoString = ser.readline().decode("utf-8") #.strip()
        
        dataArray = arduinoString.split(',')      # Since the values are being sent consecutively,
        # Data conversion from string to float
        Accx[-1].append(float(dataArray[0]))
        Accy[-1].append(float(dataArray[1]))
        Accz[-1].append(float(dataArray[2]))
        Gyrx[-1].append(float(dataArray[3]))
        Gyry[-1].append(float(dataArray[4]))
        Gyrz[-1].append(float(dataArray[5].split('\r\n')[0]))
        
        if len(Accx[-1]) >= min_length:
            stdAccx = np.std(Accx[-1][len(Accx[-1])-steady_hand_window:])
            stdAccy = np.std(Accy[-1][len(Accy[-1])-steady_hand_window:])
            stdAccz = np.std(Accz[-1][len(Accz[-1])-steady_hand_window:])
            if stdAccx < steady_hand_threshold and stdAccy < steady_hand_threshold and stdAccz < steady_hand_threshold:
                Accx.append([])
                Accy.append([])
                Accz.append([])
                Gyrx.append([])
                Gyry.append([])
                Gyrz.append([])
                labels.append(letra)
                count += 1
                print("OK")
                letter_counter += 1
                letra = random.choice(letters)        
                print(str(letter_counter) + " - " + letra)
    except Exception as e:
        print(e)
        print(arduinoString)        
        pass
print("THANKS")
```

```python
ser.close()
```

```python
raw_data = {'Accx': Accx[2:-1], 'Accy': Accy[2:-1], 'Accz': Accz[2:-1], 'Gyrx': Gyrx[2:-1], 'Gyry': Gyry[2:-1],
            'Gyrz': Gyrz[2:-1], 'labels': labels[2:]}
```

```python
with open("raw_data_marcelo_abcde_25022019_2337.json", 'w') as f:
    json.dump(raw_data, f)
```
